```r
# Load libraries once
library(Seurat)
library(tidyverse)
library(cowplot)
library(patchwork)
library(devtools)
library(WGCNA)
library(hdWGCNA)

# Using the cowplot theme for ggplot
theme_set(theme_cowplot())

# Set random seed for reproducibility
set.seed(12345)



# Define the function outside the loop
# Define the function outside the loop
microglia_script <- function(cellstate) {
    seurat_obj <- readRDS("/users/PAS2694/eblige2/allCellTypes_garnettCleaned_finalObj_294114nuclei.rds")
    seurat_obj <- UpdateSeuratObject(object = seurat_obj)
    DefaultAssay(seurat_obj) <- "RNA"
    seurat_obj <- subset(seurat_obj, subset = cellState == cellstate)
    
    # Analysis and plotting steps
    p <- DimPlot(seurat_obj, group.by = 'cellState', label = TRUE) +
         umap_theme() + ggtitle('oligo') + NoLegend()
    seurat_obj <- SetupForWGCNA(seurat_obj, gene_select = "fraction", fraction = 0.05, wgcna_name = cellstate)
    seurat_obj <- MetacellsByGroups(seurat_obj = seurat_obj, group.by = c("Sample_ID", 'cellType'), reduction = 'pca', k = 36, max_shared = 10, ident.group = 'cellType', min_cells = 100)
    seurat_obj <- NormalizeMetacells(seurat_obj)
    DefaultAssay(seurat_obj) <- "RNA"
    
    seurat_obj <- SetDatExpr(seurat_obj, group_name = "Oligo", group.by = 'cellType', assay = 'RNA', slot = 'data')
    seurat_obj <- TestSoftPowers(seurat_obj, networkType = 'signed')
    plot_list <- PlotSoftPowers(seurat_obj)
    wrap_plots(plot_list, ncol = 2)
    
    # Construct the network and save the TOM to a uniquely named CSV
    tom_name <- paste0("TOM_", cellstate)
    seurat_obj <- ConstructNetwork(seurat_obj, tom_name = tom_name, overwrite_tom = TRUE)
    tom <- seurat_obj@misc[[tom_name]]  # Assuming the TOM is stored here
    write.csv(tom, paste0(tom_name, ".csv"))  # Save the TOM with a unique name
    
    PlotDendrogram(seurat_obj, main = paste(cellstate, 'hdWGCNA Dendrogram'))
    seurat_obj <- ScaleData(seurat_obj, features = VariableFeatures(seurat_obj))
    seurat_obj <- ModuleEigengenes(seurat_obj, group.by.vars = "Sample_ID")
    seurat_obj <- ModuleConnectivity(seurat_obj, group.by = 'cellType', group_name = 'Oligo')
    
    modules <- GetModules(seurat_obj) %>% subset(module != 'grey')
    hub_df <- GetHubGenes(seurat_obj, n_hubs = 30)
    MEs <- GetMEs(seurat_obj, harmonized = TRUE)
    seurat_obj@meta.data <- cbind(seurat_obj@meta.data, MEs)
    
    seurat_obj <- ResetModuleNames(seurat_obj, new_name = "oligo")
    plot_list <- ModuleFeaturePlot(seurat_obj, features = 'hMEs', order = TRUE)
    wrap_plots(plot_list, ncol = 6)
    
    cell_types <- seurat_obj@meta.data$cellType
    gene_expression <- FetchData(seurat_obj, vars = "PSEN1")
    module_eigengenes <- seurat_obj@reductions$pca@cell.embeddings
    correlations <- cor(gene_expression, module_eigengenes)
    write.csv(correlations, paste0("ME_cor_", cellstate, ".csv"))
    
    gene_expression_by_cell_type <- aggregate(FetchData(seurat_obj, vars = "PSEN1"), by = list(cell_types), FUN = mean)
    print(gene_expression_by_cell_type)
    
    VlnPlot(seurat_obj, features = "PSEN1", group.by = "cellType")
    modules <- GetModules(seurat_obj) %>% subset(module != 'grey')
    
    return(seurat_obj)
}

# Initialize an empty list to store results
results <- list()

# Loop through each state in cellstatelist
for (i in seq_along(cellstatelist)) {
    state <- cellstatelist[i]
    results[[paste0("obj", i)]] <- microglia_script(state)
}

# Now `results` contains all Seurat objects created for each cell state
```
