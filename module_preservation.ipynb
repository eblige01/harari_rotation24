```r
# Function to perform module preservation between each pair of Seurat objects
run_module_preservation <- function(query_obj, ref_obj, name_prefix, n_permutations = 250) {
  # Module preservation analysis between query and reference Seurat objects
  result <- ModulePreservation(
    query_obj,
    ref_obj,
    name = paste(name_prefix, "comparison"),
    n_permutations = n_permutations,
    parallel = FALSE,
    seed = 12345,
    return_raw = TRUE
  )
  return(result)
}
```

```r
# Run the module preservation analysis for each pair (query vs. reference)
preservation_results <- list()

for (i in 1:length(seurat_objs_query)) {
  preservation_results[[paste0("state_", i)]] <- run_module_preservation(
    seurat_objs_query[[i]],  # Query Seurat object
    seurat_objs_ref[[i]],    # Reference Seurat object
    name_prefix = paste("state", i),
    n_permutations = 100
  )
}
```
